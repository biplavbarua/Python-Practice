```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#Load the Titanic Dataset
titanic = sns.load_dataset('titanic')
```

```python
#Data Exploration (EDA)
print(titanic.head())
print(titanic.info())
print(titanic.describe())

#Missing Values
print(titanic.isnull().sum())
```

```python
#Viusalization
#1. How does passenger age distribution look like?
#2. What is the survival rate across different classes? (using barplot)
#3. How does fare vary with passenger class? (using boxplot)
```

```python
#Data Visualization: Heatmap
plt.figure(figsize=(8, 5))
sns.countplot(x='survived', data=titanic, palette='coolwarm')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Q1. How does passenger age distribution look like?

plt.figure(figsize=(8, 6))
plt.hist(titanic['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Q2. What is the survival rate across different classes? (using barplot)

plt.figure(figsize=(8, 6))
sns.barplot(x='class', y='survived', data=titanic, color='red', edgecolor='green', errorbar=None)
plt.title('Survival Rate Across Different Classes')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#Q3. How does fare vary with passenger class? (using boxplot)

plt.figure(figsize=(8, 6))
sns.boxplot(x='class', y='fare', data=titanic, hue='class')
plt.title('Fare Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```
